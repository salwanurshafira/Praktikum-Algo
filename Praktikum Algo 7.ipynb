```python
"""
Nama = Salwa Nur Shafira
NIM = 065002500019
Tanggal = Kamis, 13 November 2025
Tugas = Menentukan bilangan prima atau bukan
"""

def cek_prima(x):
    if x <= 1:
        return False
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

def tampil_hasil_prima(angka, status_prima):
    if status_prima:
        print(f"{angka} adalah bilangan prima.")
    else:
        print(f"{angka} bukan bilangan prima.")

angka_input = int(input("Masukkan angka: "))
status_prima = cek_prima(angka_input)
tampil_hasil_prima(angka_input, status_prima)
```

```python
"""
Nama = Salwa Nur Shafira
NIM = 065002500019
Tanggal = Kamis, 13 November 2025
Tugas = Ordinal Number
"""

def dapatkan_akhiran_ordinal(n):
    if n == 0:
        return 'th'

    if 10 <= n % 100 <= 20:
        akhiran = 'th'
    else:
        digit_terakhir = n % 10
        if digit_terakhir == 1:
            akhiran = 'st'
        elif digit_terakhir == 2:
            akhiran = 'nd'
        elif digit_terakhir == 3:
            akhiran = 'rd'
        else:
            akhiran = 'th'

    return akhiran


print("Ordinal Number")
print("Ketik 0 untuk menghentikan program")

while True:
    angka = int(input("Masukkan angka: "))

    if angka == 0:
        print("Terima kasih telah menggunakan program saya.")
        break

    suffix = dapatkan_akhiran_ordinal(angka)

    print(f"({angka}, '{suffix}')")
```
